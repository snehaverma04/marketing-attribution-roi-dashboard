# Week 1 - Exploratory Data Analysis (EDA)

## Objective

The objective of this notebook is to analyze the marketing datasets and identify trends, patterns, and insights that can help evaluate campaign performance and user conversions.

### Analysis to Perform

- Channel-wise analysis
- Campaign-wise analysis
- Conversion analysis
- Time-based analysis


```python
import pandas as pd
import matplotlib.pyplot as plt

# Display all columns
pd.set_option("display.max_columns", None)

# Load datasets
attribution = pd.read_csv("../data/raw/multi_touch_attribution_data.csv")
performance = pd.read_csv("../data/raw/tech_advertising_campaigns_dataset.csv")

print("Datasets loaded successfully!")
```

```python
# Standardize column names

attribution.columns = (
    attribution.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
# Campaign distribution

attribution["campaign"].value_counts()
```

```python
# Clean campaign values

attribution["campaign"] = attribution["campaign"].replace("-", "Unknown")
```

```python
attribution["campaign"].value_counts().plot(kind="bar")
plt.title("Campaign-wise User Distribution")
plt.xlabel("Campaign")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.show()
```

### Observation

- Unknown campaign has the highest number of users.
- Other campaigns have a relatively similar user distribution.


```python
# Conversion distribution

attribution["conversion"].value_counts()
```

```python
# Conversion-wise distribution

attribution["conversion"].value_counts().plot(kind="bar")
plt.title("Conversion Distribution")
plt.xlabel("Conversion Status")
plt.ylabel("Number of Users")
plt.xticks(rotation=0)
plt.show()
```

```python
# Convert timestamp to datetime

attribution["timestamp"] = pd.to_datetime(attribution["timestamp"])

# Create hour column if not present

attribution["hour"] = attribution["timestamp"].dt.hour

# Hour-wise user distribution

attribution["hour"].value_counts().sort_index()
```

```python
attribution["hour"].value_counts().sort_index().plot(kind="bar")

plt.title("Hour-wise User Distribution")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Users")
plt.show()
```

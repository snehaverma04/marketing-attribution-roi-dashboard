```python
import pandas as pd

attribution = pd.read_csv("../data/raw/multi_touch_attribution_data.csv")

print(attribution.columns)
```

```python
# Standardize column names

attribution.columns = (
    attribution.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

print(attribution.columns)
```

```python
# Convert timestamp column to datetime

attribution["timestamp"] = pd.to_datetime(attribution["timestamp"])

print("Timestamp converted successfully!")
```

```python
attribution.info()
```

```python
# Create date-related features

attribution["date"] = attribution["timestamp"].dt.date
attribution["year"] = attribution["timestamp"].dt.year
attribution["month"] = attribution["timestamp"].dt.month
attribution["day"] = attribution["timestamp"].dt.day
attribution["hour"] = attribution["timestamp"].dt.hour
attribution["day_of_week"] = attribution["timestamp"].dt.day_name()

print("Date features created successfully!")
```

```python
attribution["conversion"] = attribution["conversion"].str.lower()
```

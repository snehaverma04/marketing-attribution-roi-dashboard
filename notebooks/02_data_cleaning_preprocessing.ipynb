# Week 1 - Data Cleaning & Preprocessing

## Objective

The objective of this notebook is to clean, standardize, and prepare the marketing datasets for analysis.

### Tasks

- Load datasets
- Standardize column names
- Convert timestamp to datetime
- Handle campaign values
- Convert conversion labels
- Save cleaned datasets

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
# Load Attribution Dataset
attribution = pd.read_csv("../data/raw/multi_touch_attribution_data.csv")

# Load Advertising Performance Dataset
performance = pd.read_csv("../data/raw/tech_advertising_campaigns_dataset.csv")

print("Datasets Loaded Successfully")
```

```python
# Standardize Attribution Dataset Column Names

attribution.columns = (
    attribution.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

# Standardize Performance Dataset Column Names

performance.columns = (
    performance.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

```

```python
print("Attribution Columns:")
print(attribution.columns)

print("\nPerformance Columns:")
print(performance.columns)
```

```python
# Convert timestamp to datetime

attribution["timestamp"] = pd.to_datetime(attribution["timestamp"])

```

```python
attribution.info()
```

```python
# Standardize conversion values

attribution["conversion"] = (
    attribution["conversion"]
    .str.strip()
    .str.lower()
)
```

```python
attribution["conversion"].value_counts()
```

```python
# Replace invalid campaign values

attribution["campaign"] = attribution["campaign"].replace("-", "Unknown")

print("Campaign values cleaned successfully.")
```
